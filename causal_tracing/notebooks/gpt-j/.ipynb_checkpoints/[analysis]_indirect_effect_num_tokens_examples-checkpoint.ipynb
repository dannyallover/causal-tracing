```python
import torch
import random

import sys
sys.path.append("../../../lib")
import utility
```

```python
res = torch.load("../../data/gpt-j/indirect_effect_1000_examples.pt")
```

```python
# 100 examples with least tokens
for prob, access in res[-100:]:
    text, tkens, num_layers, _, _ = access
    x = num_layers
    y = tkens
    bound = max(torch.max(prob).item(), abs(torch.min(prob).item()))
    incr = bound / 10000
    token = text.split()[-1]
    title = f"p*,h({token}) - p*({token})"
    color_schema = "PuOr"
    
    utility.plot_results(prob, x, y, -bound, bound, incr, title, color_schema)
```

```python
# 100 examples with most tokens
for prob, access in res[:100]:
    text, tkens, num_layers, _, _ = access
    x = num_layers
    y = tkens
    bound = max(torch.max(prob).item(), abs(torch.min(prob).item()))
    incr = bound / 10000
    token = text.split()[-1]
    title = f"p*,h({token}) - p*({token})"
    color_schema = "PuOr"
    
    utility.plot_results(prob, x, y, -bound, bound, incr, title, color_schema)
```
