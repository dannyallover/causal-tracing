```python
import torch

import sys
sys.path.append("../../../lib")
import utility
```

```python
res = torch.load("../../data/gpt-j/indirect_effect_1000_examples.pt")
```

```python
num_layers = res[0][1][2]
sd = utility.standard_deviation_sites(res, num_layers)
```

```python
x = num_layers
y = ["first subject token", "middle subject tokens", "last subject token", \
              "first subsequent token", "further tokens", "last token"]
bound = max(torch.max(sd).item(), abs(torch.min(sd).item()))
incr = bound / 10000
title = "Standard deviation"
color_schema = "PuOr"

utility.plot_results(sd, x, y, -bound, bound, incr, title, color_schema)
```
