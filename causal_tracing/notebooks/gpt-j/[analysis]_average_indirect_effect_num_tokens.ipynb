```python
import torch
import random

import sys
sys.path.append("../../../lib")
import utility
```

```python
res = torch.load("../../data/gpt-j/indirect_effect_1000_examples.pt")
```

```python
num_layers = res[0][1][2]
avg_IE_lowest_num_tokens = utility.average_sites(res[-100:], num_layers)
avg_IE_highest_num_tokens = utility.average_sites(res[:100], num_layers)
```

```python
# zero cutoff
x = num_layers
y = ["first subject token", "middle subject tokens", "last subject token", \
              "first subsequent token", "further tokens", "last token"]
lower_bound = 0
upper_bound = torch.max(avg_IE_lowest_num_tokens).item()
incr = (upper_bound - lower_bound) / 10000
title = "AIE"
color_schema = "Purples"

utility.plot_results(avg_IE_lowest_num_tokens, x, y, lower_bound, upper_bound, incr, title, color_schema)
```
